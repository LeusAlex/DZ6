```python
import json

# Открываем и читаем файл purchase_log.txt
with open('purchase_log.txt', 'r', encoding='utf-8') as file:
    content = file.readlines()

# Удаляем заголовок файла
content.pop(0)

# Создаем пустой словарь
purchases = {}

# Проходим по каждой строке содержимого файла
for line in content:
    # Преобразуем строку в словарь
    data = json.loads(line)
    # Извлекаем значение ключа 'user_id'
    user_id = data['user_id']
    # Извлекаем значение ключа 'category'
    category = data['category']
    # Добавляем пару ключ-значение в словарь purchases
    purchases[user_id] = category

# Выводим первые два элемента словаря purchases
for user_id, category in list(purchases.items())[:-1]:
    print(user_id, category)
```
